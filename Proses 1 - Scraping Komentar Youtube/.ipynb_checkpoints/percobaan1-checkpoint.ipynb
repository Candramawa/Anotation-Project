```python
from selenium import webdriver
import time
```

```python
driver = webdriver.Chrome()
```

```python
url = "https://www.youtube.com/watch?v=rQX3CzUyu58"
```

```python
driver.get(url)
print(driver.current_url)
driver.implicitly_wait(30)
```

```python
#BUAT SCROLL OTOMATIS TAPI BELUM JALAN

for i in range(1,7):
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    # beri waktu jeda untuk load item anime
    # perhatikan perubahan di google chrome, halaman akan scroll down otomatis
    time.sleep(20)
```

```python
results = driver.find_elements_by_id("comment")
```

```python
len(results)
```

```python
for komentar in results:
    print(komentar.text)
    print("================")
```

```python
komen = []
```

```python
for komentar in results:
    isi = komentar.find_element_by_xpath("//*[@id='main']")
    text = komentar.find_element_by_xpath("//*[@id='content-text']").text
    #isi = komentar.find_element_by_id("main")
    c_text = isi.text.split("\n")
    komen.append({
        'isi' : text,
        'like': c_text[4]
    })
```
